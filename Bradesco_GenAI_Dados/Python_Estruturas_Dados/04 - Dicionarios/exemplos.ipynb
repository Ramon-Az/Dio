```python
contatos = {
    "guilherme@gmail.com": {"nome": "Guilherme", "telefone": "3333-2221"},
    "giovanna@gmail.com": {"nome": "Giovanna", "telefone": "3443-2121"},
    "chappie@gmail.com": {"nome": "Chappie", "telefone": "3344-9871"},
    "melaine@gmail.com": {"nome": "Melaine", "telefone": "3333-7761", "extra": {"a": 1}}
}

contatos["giovanna@gmail.com"]["telefone"]
```

```python
extra = contatos["melaine@gmail.com"]["extra"]["a"]
print(extra)
```

```python
# iterar dicionario

for chave in contatos:
    print(chave, contatos[chave])
```

```python
# outra forma de iterar dicionario
for chave, valor in contatos.items():
    print(F"chave: {chave}, valor: {valor}")
```

```python
# métodos da classe dict

copia = contatos.copy()
print(copia)
```

```python
copia["guilherme@gmail.com"] = {"nome": "GUI"}
print(copia)
```

```python
copia["guilherme@gmail.com"]
```

```python
dict.fromkeys(["nome", "telefone"]) # {"nome": None, "telefone": None}
```

```python
dict.fromkeys(["nome", "telefone"], "vazio") #{"nome": "vazio", "telefone": "vazio"}
```

```python
contatos ={
    "guilherme@gmail.com": {
        "nome": "Guilherme", 
        "telefone": "3333-2221"
    }
}

contatos["chave"]
```

```python
resultado = contatos.get("chave")
print(resultado)
```

```python
resultado = contatos.get("chave", {})
print(resultado)
```

```python
# get retorna erro se não encontrar a chave, o erro de acordo com o que foi definido
resultado = contatos.get("guilherme@gmail.com", {})
print(resultado)
```

```python
novo_dicionario = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 5, "f": 6, "g": 7, "h": 8}
novo_dicionario.keys()
```

```python
resultado = novo_dicionario.pop("a", "nao encontrei o elemento para remover")
print(resultado)
```

```python
novo_dicionario.setdefault("a", 1)
```

```python
novo_dicionario.setdefault("i", 100)
print(novo_dicionario)
```

```python
novo_dicionario.update({"a": {
    "novo_nome": "guilherme",
    "nova_idade": 28}})

print(novo_dicionario)
```

```python
novo_dicionario.values()
```

```python
resultado = "novo_nome" in novo_dicionario["a"]
print(resultado)
```

```python
# del remove um item do dicionário, até mesmo o proprio dicionarios
del novo_dicionario["a"]
print(novo_dicionario)
```
