```python
numeros = [1, 30, 21, 2, 9, 65, 34]
pares = []

for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)

print(f"Os numeros pares da lista: {numeros} são: {pares}")
```

```python
# comprehend
numeros = [1, 30, 21, 2, 9, 65, 34]
pares =[numero for numero in numeros if numero % 2 == 0]

print(pares)
```

```python
quadrado = [
    numero ** 2 for numero in numeros
]

print(f"O quadrado dos numeros da Lista: {numeros} são {quadrado}")
```

```python
linguagens = [
    "python",
    "js",
    "c"
]

print(linguagens)
```

```python
# .extend usado para passar uma lista de elementos para outra lista já existente, adicionando os novos valores ao final.
linguagens.extend(["java", "csharp"])
print(linguagens)
```

```python
linguagens.pop()
print(linguagens)
```

```python
linguagens.pop(2)
print(linguagens)
```
