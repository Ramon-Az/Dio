Estruturas condicionais e de repetição

```python
def sacar(valor, saldo):
    if valor <= saldo:
        saldo -= valor
        print(f"Saque de R${valor:.2f} realizado com sucesso. Saldo atual: R${saldo:.2f}")
    else:
        print("Saldo insuficiente para realizar o saque.")
    return saldo

saldo = 1000
saldo = sacar(200, saldo)

print("Obrigado por usar nosso cliente! Tenha um ótimo dia!")
```

```python
def depositar(valor, saldo):
    if valor > 0:
        saldo += valor
        print(f"Depósito de R${valor:.2f} realizado com sucesso. Saldo atual: R${saldo:.2f}")
    else:
        print("Valor de depósito deve ser positivo.")
    return saldo

saldo = depositar(500, saldo)
print("Obrigado por usar nosso cliente! Tenha um ótimo dia!")
```

```python
MAIOR_IDADE = 18
IDADE_ESPECIAL = 12

idade = int(input("Digite sua idade: "))

if idade >= MAIOR_IDADE:
    print("Maior de idade, pode tirar a CNH.")

if idade < MAIOR_IDADE:
    print("Menor de idade, não pode tirar a CNH.")
```

```python
if idade >= MAIOR_IDADE:
    print("Maior de idade, pode tirar a CNH.")
else:
    print("Menor de idade, não pode tirar a CNH.")
```

```python
if idade >= MAIOR_IDADE:
    print("Maior de idade, pode tirar a CNH.")
elif idade == IDADE_ESPECIAL:
    print("Pode fazer aulas teóricas, mas não pode fazer aulas práticas.")
else:
    print("Menor de idade, não pode tirar a CNH.")
```

```python
conta_normal = True
conta_univertaria = False

saldo = 2000
saque = 500
cheque_especial = 500

if conta_normal:
    if saldo >= saque:
        print("Saque realizado com sucesso.")
    elif saque <= (saldo + cheque_especial):
        print("Saque realizado com sucesso utilizando o limite.")
    else:
        print("Saldo insuficiente!.")
        
elif conta_univertaria:
    if saldo >= saque:
        print("Saque realizado com sucesso.")
    else:
        print("Saldo insuficiente!.")
else:
    print("O sistema não reconhece o tipo de conta, entre em contato com seu gerente!.")
```

```python
# if intermediario

status = "sucesso" if saldo >= saque else "falha"

print(f"{status} ao realizar o saque.")
```
