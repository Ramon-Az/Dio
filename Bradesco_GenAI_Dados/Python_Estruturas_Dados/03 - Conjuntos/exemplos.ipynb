```python
#converter o set para uma lista
numeros ={1, 2, 3, 2}
print(numeros)
```

```python
#converter o set para uma lista, final formato 1
numeros = list(numeros)
print(numeros[0])
```

```python
#converter o set para uma lista, final formato 2, interar dentro de um for
numeros = {1, 2, 3, 2}

for i in numeros:
    print(i)
```

```python
# funcao enumerate, para saber o indice do conjunto
carros = {"gol", "celta", "palio"}

for indice, carro in enumerate(carros):
    print(f"indice {indice}: {carro}")
```

```python
#metodos da classe set

conjunto_a = {1,2,3}
conjunto_b = {2,3,4}

conjunto_a.union(conjunto_b)
```

```python
conjunto_a.intersection(conjunto_b)
```

```python
conjunto_a.difference(conjunto_b)
```

```python
conjunto_b.difference(conjunto_a)
```

```python
conjunto_a.symmetric_difference(conjunto_b)
```

```python
conjunto_1 = {1, 2, 3}
conjunto_2 = {4, 1, 2, 5, 6, 3}

conjunto_1.issubset(conjunto_2)
```

```python
conjunto_2.issubset(conjunto_1)
```

```python
conjunto_1.issuperset(conjunto_2)
```

```python
conjunto_1.isdisjoint(conjunto_2)
```
