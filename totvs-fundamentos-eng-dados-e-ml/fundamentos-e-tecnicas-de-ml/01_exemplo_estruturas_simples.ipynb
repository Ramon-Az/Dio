**Estruturas condicional simples**

```python
numero_1, numero_2 = 1, 2
print(numero_1 > numero_2)
```

```python
print(numero_1)
```

```python
print(numero_2)
```

```python
if numero_1 <= numero_2:
    print('Sim')
```

**Estrutura condicional aninhada**

```python
soma = numero_1 + numero_2
if soma >= 0:
    print('Sim, a soma é maior ou igual a zero')
elif soma == 0:
    print('A soma é igual a zero')    
else:
    print('Não, a soma é menor que zero')
```

**Estruturas de repetição**

```python
# loop com for
lista = [
    'Ramon',
    'Patricia',
    'Leticia',
    'Carlos'
]

for i in lista:
    print(i)
```

```python
# loop com while
contador = 0

while contador < 4:
    print(lista[contador])
    contador += 1
```
